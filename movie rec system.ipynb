```python
def convert3(obj):
    L = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter < 3:
            L.append(i['name'])
        counter+=1
    return L 
```

```python
movies['cast'] = movies['cast'].apply(convert3)
```

```python
# now everything is in list except 'overview' i.e. string, and we need our colounns to be list like others.
movies['overview'][0]
```

```python
# for crew, we would only extract the name where job is director since we only need it.
def fetch_director(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L 
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies.head()
# now everything is in list and we will concatinate all coloumns to become a large list and then we will convert the list into string to become a large paragraph which is ultimately the tags coloumn.
# but there is one problem, removal of space
```

```python
def revspace(obj):
    L1 = []
    for i in obj:
        L1.append(i.replace(" ",""))
    return L1
```

```python
movies['cast'].apply(revspace)
```

```python
# applying this on all 4 coloumns
movies['cast'] = movies['cast'].apply(revspace)
movies['crew'] = movies['crew'].apply(revspace)
movies['genres'] = movies['genres'].apply(revspace)
movies['keywords'] = movies['keywords'].apply(revspace)
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):                            #creating a helper function
    y=[]
    
    for i in text.split():                 #converting string into list
        y.append(ps.stem(i))               #steming on each word   
        
    return " ".join(y)
```

```python
new_df['tags'].apply(stem)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
#created an object cv taking 2 parameters
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
#converting explicitly into a numpy array because cv returns a sparse matrix by default and we need a numpy array
cv.fit_transform(new_df['tags']).toarray()
```

```python
cv.fit_transform(new_df['tags']).toarray().shape
```

```python
#storing in 'vectors' variable
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
#pass the vectors in the function
cosine_similarity(vectors)
```

```python
#saving the value in "similarity" variable
similarity = cosine_similarity(vectors)
```

```python
#creating a function which returns 5 recommended movies when one movie is input
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
    
    for i in movies_list:
        print(i[0])
```
